# Display grayscale image #

```python
# Display a gray scale image
# Change this to load different images
import cv2
img = cv2.imread('images/cameraManGray.jpg')
cv2.imshow('the image', img) 
cv2.waitKey() 
cv2.destroyAllWindows()
```

```python
# Check pixel value with opencv
plt.imshow(img)
plt.show()

# check shape of matrix 
print(img.shape) # (225, 225, 3) but gray value = [13, 13, 13]

# convert to 2D 
G = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(G.shape) # (225, 225 )(image resolution)

 # Access a single pixel, show colour in gray value
print( G[150, 50] )      # 13, opencv use coordinate as [row index, column index] 
```

```python
print(G)
print('\n')
print(G[0,0], G[0,224])
```

# Display colour image #

```python
# Display an image
# Change this to load different images
img = cv2.imread('images/myImage.jpg')
cv2.imshow('the image', img) 
cv2.waitKey() 
cv2.destroyAllWindows()
```

```python
# easier to use matplotlib to plot if use Jupyter
# open cv use BGR format , matplotlib expect RGB
img2=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.show()

 # Access a single pixel,[34, 40,  36] , show colour in RGB format
img2[150, 50, :]       # opencv use coordinate as [row index, column index]   
```

```python
# Convert image to gray scale
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('Original image',img)
cv2.imshow('Gray image', imgGray)
  
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
type(imgGray)
print(imgGray.shape) # (368, 491)
print(img.shape) # (368, 491, 3)
```

```python
# split colour image to 3 separate channel
plt.imshow(img2)
f, axes = plt.subplots(1,3)
i=0
for ax in axes:
    ax.imshow(img2[:,:,i], cmap='gray') #RGB
    i +=1
```

```python
# Create a green image
# define image with height=240, width=320, 3 channels

import numpy as np
shape = (240, 320, 3)
im = np.empty(shape)
im[:] = (0, 255, 0) # RGB format
plt.imshow(im)
plt.show()
im
```

```python
img = cv2.imread('images/myImage.jpg')
cv2.imshow('the image', img) 

im_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow('hsv image', im_hsv)

cv2.waitKey(0) 
cv2.destroyAllWindows()

plt.imshow(im_hsv, cmap='hsv')
plt.show()
```
